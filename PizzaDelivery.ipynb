```python
def secim():
    sos_fatura = 0
    taban_fatura = 0
    secim_sos = []
    secimler = []
    
    soslar = {1: 'Zeytin', 2: 'Mantar', 3: 'Keçi Peyniri', 4: 'Et', 5: 'Soğan', 6: 'Mısır'}
    sos_fiyat = {'Zeytin':5, 'Mantar':10, 'Keçi Peyniri':15, 'Et':15, 'Soğan':5, 'Mısır':5}
    
    tabanlar = {1: 'Klasik', 2: 'Margarita', 3: 'Türk Pizza', 4: 'Sade Pizza'}
    taban_fiyat = {'Klasik':30, 'Margarita':40, 'Türk Pizza':25, 'Sade Pizza':20}
    
    """Fonksiyonda kullanmak üzere soslar ve pizza tabanları için ayrı ayrı fatura değişkenleri oluşturdum
        ve bu değişkenleri 0'a eşitledim.
        Aynı şekilde ileride kullanmak üzere iki adet liste oluşturdum. Daha sonra ekrana basmakta ve seçim
        yapılmasını kolaylaştırmak için 'soslar' ve 'tabanlar' sözlüklerini, ücret hesabı yapmayı kolaylaştırmak için de
        'sos_fiyat' ve 'taban_fiyat' sözlüklerini oluşturdum."""
    
    """Müşteriye kaç adet pizza tabanı istediğini sordum. 0 ile programdan çıkmak mümkün. 1 değeri girilirse for döngüsü
        1 kez çalışıyor. 10 değeri girilirse 10 kez çalışır.
        Ardından if-else yapısı kurdum ve seçilen pizza tabanlarını önceden oluşturduğum 'secimler' listesine yolladım
        ve ekrana seçilen tabanı listeye erişerek bastırdım.
        Örneğin: Müşteri tek bir pizza istemiş olsun, o da 'Klasik' olsun. Bunu append ile listeye yolladığımda,
        'secimler' listesi şu hale gelir: secimler = ["Klasik"]. i=1 olur, secimler[i-1] de secimler[0] olur ve
        listenin elemanı ekrana basılmış olur.
        Daha sonra, pizza seçimini sağlayan her bir elif bloğunun altına sos seçimi opsiyonu koydum. Birden çok sos
        opsiyonu sağlayabilmek için .split() kullandım. input() ile direkt string olarak aldım, ardından da girilen
        her sayı değerini int'e çevirdim."""
    
    taban_sayisi = int(input("Kaç adet pizza istiyorsunuz? 0 harici bir sayı giriniz: "))
    
    
    for i in range(1, taban_sayisi+1):
        
        secim_taban = int(input(str(i)+". pizza seçimi: \
                                \nKlasik (30TL) için 1\
                                \nMargarita (40 TL) için 2 \
                                \nTürk Pizza (25 TL) için 3 \
                                \nSade Pizza için (20 TL) 4 tuşuna basınız. \
                                \nProgramdan çıkmak için 0 tuşuna basabilirsiniz. "))
    
        if secim_taban == 0:
            print("\nProgramdan çıkılıyor...")
            exit()
    
        elif secim_taban == 1:
            secimler.append("Klasik")
            print("\n"+str(i)+". pizzanız için taban seçiminiz: ",secimler[i-1])
            
            secim_sos = input(str(i)+". pizza için sos seçimi: \
                              \nZeytin (5 TL) için 1 \
                              \nMantar (10TL) için 2 \
                              \nKeçi peyniri (15 TL) için 3 \
                              \nEt (15 TL) için 4 \
                              \nSoğan (5 TL) için 5\
                          \nMısır (5 TL) için 6 tuşuna basınız. \
                         \nBoşluk bırakarak çoklu seçim yapabilirsiniz. Örn: 1 2 3. ").split()
            
            secim_sos = [int(secim) for secim in secim_sos]
                          
    
        elif secim_taban == 2:
            secimler.append("Margarita")
            print("\n"+str(i)+". pizzanız için taban seçiminiz: ",secimler[i-1])
            
            secim_sos = input(str(i)+". pizza için sos seçimi: \
                              \nZeytin (5 TL) için 1 \
                              \nMantar (10TL) için 2 \
                              \nKeçi peyniri (15 TL) için 3 \
                              \nEt (15 TL) için 4 \
                              \nSoğan (5 TL) için 5\
                          \nMısır (5 TL) için 6 tuşuna basınız. \
                         \nBoşluk bırakarak çoklu seçim yapabilirsiniz. Örn: 1 2 3. ").split()
                        
            secim_sos = [int(secim) for secim in secim_sos]
    
        elif secim_taban == 3:
            secimler.append("Türk Pizza")
            print("\n"+str(i)+". pizzanız için taban seçiminiz: ",secimler[i-1])
            
            secim_sos = input(str(i)+". pizza için sos seçimi: \
                              \nZeytin (5 TL) için 1 \
                              \nMantar (10TL) için 2 \
                              \nKeçi peyniri (15 TL) için 3 \
                              \nEt (15 TL) için 4 \
                              \nSoğan (5 TL) için 5\
                          \nMısır (5 TL) için 6 tuşuna basınız. \
                         \nBoşluk bırakarak çoklu seçim yapabilirsiniz. Örn: 1 2 3. ").split()
                          
            secim_sos = [int(secim) for secim in secim_sos]
    
        elif secim_taban == 4:
            secimler.append("Sade Pizza")
            print("\n"+str(i)+". pizzanız için taban seçiminiz: ",secimler[i-1])
            
            secim_sos = input(str(i)+". pizza için sos seçimi: \
                              \nZeytin (5 TL) için 1 \
                              \nMantar (10TL) için 2 \
                              \nKeçi peyniri (15 TL) için 3 \
                              \nEt (15 TL) için 4 \
                              \nSoğan (5 TL) için 5\
                          \nMısır (5 TL) için 6 tuşuna basınız. \
                         \nBoşluk bırakarak çoklu seçim yapabilirsiniz. Örn: 1 2 3. ").split()
                         
            secim_sos = [int(secim) for secim in secim_sos]
        
        else:
            print("Yanlış bir giriş yaptınız. Tekrar deneyiniz.")
            
            
        """'secim_sos' listesinin icindeki her elemanı - secim_sos listesi burada müşterinin seçtiği sosların
            int değerleridir, örn: [1, 2, 3] - yukarıda tanımladığım sos sözlüğünün anahtarlarıyla eşit olduğunda
            müşterinin seçtiği sosları ekrana bastırdım. Daha önemlisi, pizzaları ve sosları tek bir listeye attım.
            
            Yani: secim_sos = [1, 2, 3, 4] olsun.
            soslar = {1: 'Zeytin', 2: 'Mantar', 3: 'Keçi Peyniri', 4: 'Et', 5: 'Soğan', 6: 'Mısır'} idi.
            'secim_sos' icindeki degerler 'soslar' sözlüğünün anahtarları ile aynı ise (1, 2, 3, 4, 5, 6) bunlar müşterinin
            seçtiği soslar anlamına gelmektedir, ve bunları ekrana bastırdım.
            Bunun ardından, ekrana bastırdığım değer, örneğin 'Zeytin', 
            sos_fiyat = {'Zeytin':5, 'Mantar':10, 'Keçi Peyniri':15, 'Et':15, 'Soğan':5, 'Mısır':5}
            bu sözlüğün anahtarı ile aynı ise sos fatura değerini bu anahtarın değeri kadar arttırdım.
            
            Ardından fatura değeri için aynı işlemleri pizzalar için de gerçekleştirdim ve bunları ekrana bastırdım."""
        
        print("\n******")
        print(str(i)+". pizza seçimi: ", secimler[i-1])
        print(secimler[i-1]+" pizza seçiminin yanına seçilen soslar: ")
        
        for i in secim_sos:
            if i in soslar.keys():
                print(soslar[i])
                secimler.append(soslar[i])
                if soslar[i] in sos_fiyat.keys():
                    sos_fatura += sos_fiyat[soslar[i]]
        
        for i in secimler:
            if i in tabanlar.values():
                taban_fatura += taban_fiyat.get(i)
                
    print("\nPizza tabanı seçimlerinizden faturanız:", taban_fatura)
    print("Sos seçimlerinizden faturanız: ", sos_fatura)
    print("Toplam faturanız: ",taban_fatura+sos_fatura)
    
    """ Burada while dönüsü olduğu için 'gecis_flag' değişkenine False değerini atadım. Müşteriden işleme devam etmek
        istiyorsa 1'i, çıkmak istiyorsa 0'ı tuşlamasını istedim.
        1'e basılırsa hesap fonksiyonu çağırılacaktır. 0'a basılırsa programdan çıkılacaktır. Başka bir tuşa basılırsa
        1'e veya 0'a basılana kadar döngü devam edecektir."""    
    
    gecis_flag = True
    gecis = int(input("\nÖdemeye geçmek için 1'e, programdan çıkmak için 0 tuşuna basınız. "))
        
    while gecis_flag:
        if gecis == 1:
            hesap(secimler,(taban_fatura+sos_fatura))
            gecis_flag = False
        elif gecis == 0: 
            print("Programdan çıkılıyor...")
            exit()
        else:
            print("Yanlış bir giriş yaptınız. Tekrar deneyiniz. ")
```

```python
def hesap(secim_pizza, para):
    tc = True
    kredi_karti = True
    secimler = secim_pizza
    
    isim = input("İsminiz: ")
    
    while tc:
        TC = input("TC: ")
    
        if len(TC) != 11:
            print("TC kimlik no'nuz 11 haneli olmalı. Tekrar deneyiniz. ")
        elif TC.isdigit() == False:
            print("TC kimlik no'nuz tamamen sayı içermelidir. Tekrar deneyiniz. ")
        else:
            tc = False
    
    while kredi_karti:
        kredi_karti_no = input("Boşluk bırakmamak kaydı ile kredi kartı numaranızı giriniz. ")
    
        if len(kredi_karti_no) != 16:
            print("\nKredi kartı numaranız 16 haneli olmalıdır. Tekrar deneyiniz. ")
        elif kredi_karti_no.isdigit() == False:
            print("\nKredi kartı numaranız tamamiyle sayı içermelidir. Tekrar deneyiniz. ")
        else:
            kredi_karti = False
    
    kredi_karti = True
    while kredi_karti:    
        kredi_karti_sifre = input("4 haneli kredi kartı şifrenizi giriniz. ")
        if len(kredi_karti_sifre) != 4:
            print("Kredi kartı şifreniz 4 haneli olmalıdır. Tekrar deneyiniz.")
        elif kredi_karti_sifre.isdigit() == False:
            print("Kredi kartı şifreniz tatamiyle sayı içerikli olmalıdır. Tekrar deneyiniz. ")
        else:
            kredi_karti = False
    
    kimlik_bilgileri = [isim, TC, kredi_karti_no, kredi_karti_sifre]
    csv_export(kimlik_bilgileri, secimler, para)
    return isim, TC, kredi_karti_no, kredi_karti_sifre

    """ Hesap fonksiyonunun parametreleri, içi pizza tabanları ve soslar ile dolu bir liste ve ücret değeridir.
        While döngüleri için 'tc' ve 'kredi_karti' değişkenlerine True değerini atadıktan sonra müşteriden
        bilgilerini istedim.
        Müşterinin ismi string'tir. TC'si 11 haneli olmalı ve yalnızca sayı içermelidir, aksi durumda döngüden çıkılmaz.
        
        Kredi kartı numarası ve şifre için de benzer durumlar geçerlidir.
        Hesap numarası 16 haneli ve yalnızca sayı, şifre de 4 haneli ve yalnızca sayı içermelidir.
        
        Daha sonra bu bilgileri bir listeye attım ve csv_export fonksiyonunu çağırdım."""
```

```python
def csv_export(liste1, liste2, para):
    date1 = datetime.datetime.now()
    date2 = str(date1.hour)+":"+str(date1.minute)+":"+str(date1.second)
    date2 = [date2]
    
    sozluk = [{"İsim":liste1[0], "TC":liste1[1], "Kredi Kartı No":liste1[2], "Kredi Kartı Şifresi":liste1[3],
               "Ücret": para, "Sipariş Detayı":liste2, "Sipariş Zamanı":date2}]
    
    
    
    filename = "Orders_Database.csv"
    df = pd.DataFrame(sozluk)
    df.to_csv(filename, encoding='utf-8-sig')

    """datetime kütüphanesi ile anlık zamanı elde ettim ve saat, dakika ve saniye bilgilerini aldım. Ardından
        saat:dakika:saniye olacak şekilde bu değişkeni listeye attım.
        
        Daha sonra bir sözlük oluşturdum ve csv dosyasında bulunması gereken tüm bilgileri bu sözlüğün içine attım.
        Burada liste1 müşterinin kimlik bilgilerini,
        liste2 pizza ve sos seçimlerini,
        para ise ödemesi gereken ücreti temsil etmektedir.
        
        Daha sonra bir dataframe oluşturdum ve bu sözlüğü bu dataframe'in içine attım.
        Ardından to_csv fonksiyonu ile bu dataframe'i csv dosyasına kaydettim. """
```
